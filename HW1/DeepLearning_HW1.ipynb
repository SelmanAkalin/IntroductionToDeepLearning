```python
# Resizing images
import matplotlib.pyplot as plt

img1 = plt.imread("image1.jpg")
img2 = plt.imread("image2.jpg")

from skimage.transform import rescale, resize, downscale_local_mean

img1 = resize(img1, (256, 256))
img2 = resize(img2, (256, 256))
```

```python
# Task-1.1
def flip_image_vertically(image):
    width, height = len(image[0]), len(image)

    flipped_image = [[0] * width for _ in range(height)]

    for y in range(height):
        for x in range(width):
            flipped_image[height - y - 1][x] = image[y][x]

    return flipped_image

newImg = flip_image_vertically(img1)

plt.imshow(newImg)
plt.show()
```

```python
# Task-1.2
def flip_image_horizontally(image):
    width, height = len(image[0]), len(image)

    flipped_image = [[0] * width for _ in range(height)]

    for x in range(width):
        for y in range(height):
            flipped_image[y][width - x - 1] = image[y][x]

    return flipped_image

newImg = flip_image_horizontally(img1)

plt.imshow(newImg)
plt.show()
```

```python
# Task-1.3
def rotate_image_left_90(image):
  height, width, channels = len(image), len(image[0]), len(image[0][0])

  rotatedImage = [[0] * width for _ in range(height)]

  for y in range(height):
    for x in range(width):
      rotatedImage[x][height - y - 1] = image[y][x]

  return rotatedImage

newImg = rotate_image_left_90(img1)

plt.imshow(newImg)
plt.show()
```

```python
# Task-1.4
def rotate_image_right_90(image):
  height, width, channels = len(image), len(image[0]), len(image[0][0])

  rotatedImage = [[0] * width for _ in range(height)]

  for y in range(height):
    for x in range(width):
      rotatedImage[width - x - 1][y] = image[y][x]

  return rotatedImage

newImg = rotate_image_right_90(img1)

plt.imshow(newImg)
plt.show()
```

```python
# Task-1.5
def resize_image_to_half(input_image):
    original_height, original_width = len(input_image), len(input_image[0])

    new_height = original_height // 2
    new_width = original_width // 2

    resized_image = [[[0, 0, 0] for _ in range(new_width)] for _ in range(new_height)]

    height_scale = original_height / new_height
    width_scale = original_width / new_width

    for y in range(new_height):
        for x in range(new_width):
            original_y = int(y * height_scale)
            original_x = int(x * width_scale)
            resized_image[y][x] = input_image[original_y][original_x]

    return resized_image

newImg = resize_image_to_half(img1)
plt.imshow(newImg)
plt.show()
```
